```python
coaster = coaster[['coaster_name',
    # 'Length', 'Speed',
    'Location', 'Status',
    # 'Opening date',
    #   'Type',
    'Manufacturer',
#     'Height restriction', 'Model', 'Height',
#        'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
#        'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
#        'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
#        'Track layout', 'Fastrack available', 'Soft opening date.1',
#        'Closing date',
#     'Opened', 
    # 'Replaced by', 'Website',
#        'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
#        'Single rider line available', 'Restraint Style',
#        'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
        'latitude', 'longitude',
    'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
coaster['opening_date_clean'] = pd.to_datetime(coaster['opening_date_clean'])
```

```python
coaster = coaster.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'})
```

```python
coaster = coaster.loc[~coaster.duplicated(subset=['Coaster_Name','Location','Opening_Date'])] \
    .reset_index(drop=True).copy()
```

```python
coaster['Year_Introduced'].value_counts()
```

```python
ax = coaster['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Years Coasters Introduced',color='green')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
ax = coaster['Speed_mph'].plot(kind='hist',
                          bins=20,
                          title='Coaster Speed (mph)',color='green')
ax.set_xlabel('Speed (mph)')
```

```python
ax = coaster['Speed_mph'].plot(kind='kde',
                          title='Coaster Speed (mph)',color='green')
ax.set_xlabel('Speed (mph)')
```

```python
coaster['Type_Main'].value_counts()
```

```python
coaster.plot(kind='scatter',
        x='Speed_mph',
        y='Height_ft',
        title='Coaster Speed vs. Height',color='green')
plt.show()
```

```python
ax = sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Year_Introduced',color='green',
                data=coaster)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(coaster,
             vars=['Year_Introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce'],
            hue='Type_Main')
plt.show()
```

```python
coaster_corr = coaster[['Year_Introduced','Speed_mph',
    'Height_ft','Inversions','Gforce']].dropna().corr()
coaster_corr
```

```python
sns.heatmap(coaster_corr, annot=True)
```

```python
ax = coaster.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location',color='green')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
